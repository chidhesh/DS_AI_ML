```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import plot_tree
```

```python
data = {
'Weather':['Sunny','Sunny','Cloudy','Rain','Rain','Rain','Cloudy','Sunny','Sunny','Rain','Sunny','Cloudy','Cloudy','Rain'],

'Temperature':['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild'],

'Humidity':['High','High','High','High','Normal','Normal','Normal','High','Normal','Normal','Normal','High','Normal','High'],

'Wind':['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Weak','Weak','Strong','Strong','Weak','Strong'],

'PlayFootball':['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
}

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Display dataset
df
```

```python
df.head()
```

```python
le=LabelEncoder()

#convert each column to nnumeric form
for col in df.columns:
    df[col]=le.fit_transform(df[col])
df
```

```python
#x=input features
x=df.drop('PlayFootball',axis=1)

#y=target
y=df['PlayFootball']

#display the features
print("Features (X):")
display(y)

#DIsplay target
print(y)
```

```python
#sample 
model=DecisionTreeClassifier(criterion='entropy')

#train the model
model.fit(x,y)

print("Model Trained Successfully")
```

```python
sample =[[2,0,0,1]]

#predict result
prediction=model.predict(sample)
print("Prediction:",prediction)
```

```python
plt.figure(figsize=(18,10))

#plot decision tree
plot_tree(
    model,
    feature_names=x.columns,
    filled=True,
    rounded=True,
    fontsize=12
)

plt.title("Decision Tree using TD3 Algorithm")
plt.show()
```

```python
print("Tree Depth",model.get_depth())

print("Number of nodes", model.tree_.node_count)
```
